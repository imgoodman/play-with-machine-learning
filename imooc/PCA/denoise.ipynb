```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.empty((100,2))
X[:,0]=np.random.uniform(0.0,100.0,size=100)
X[:,1]=0.75*X[:,0]+3.0+np.random.normal(0.0,10.0,size=100)
```

```python
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=1)
pca.fit(X)
x_reduction=pca.transform(X)
x_restore=pca.inverse_transform(x_reduction)
```

```python
plt.scatter(x_restore[:,0], x_restore[:,1])
plt.show()
```

## 手写识别案例

```python
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
X=digits.data
y=digits.target
```

```python
X.shape
```

```python
y.shape
```

```python
X[0]
```

```python
noisy_X= X+ np.random.normal(0.0, 4.0, size=X.shape)
```

```python
noisy_X[0]
```

```python
example_X=noisy_X[y==0,:][:10]
```

```python
for i in range(1,10):
    num=noisy_X[y==i,:][:10]
    example_X=np.vstack([example_X, num])
```

```python
example_X.shape
```

```python
example_X[0]
```

```python
def plot_digits(data):
    fig,axes=plt.subplots(10,10,figsize=(10,10), 
                          subplot_kw={'xticks':[], 'yticks':[]}, gridspec_kw=dict(hspace=0.1, wspace=0.1))
    for i,ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(8,8), cmap='binary', interpolation='nearest', clim=(0,16))
    plt.show()

plot_digits(example_X)
```

## 可以看出，手写中带有很多的噪音

```python
"""
这里的噪音比较多
因此，这里只保存50%的信息
"""
pca=PCA(0.5)
pca.fit(noisy_X)
```

```python
pca.n_components_
```

```python
example_x_reduction=pca.transform(example_X)
```

```python
example_x_restore=pca.inverse_transform(example_x_reduction)
```

```python
plot_digits(example_x_restore)
```
